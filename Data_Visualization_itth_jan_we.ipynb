Data visualization Packages

to create various types of charts

hidden patterns from data

matplotlib, seaborn, plotly ( 3d charts )

```python
import matplotlib.pyplot as plt
```

```python
x=[2,4,6,8]
y=[7,8,12,15]
plt.plot(x,y)
plt.show()
```

Line charts are used for trends over a period of time

```python
plt.bar(x,y)
```

```python
#scatter plot
plt.scatter(x,y)
plt.xlabel('Sales')
plt.ylabel('Year')
plt.title('Sales Report')
plt.show()
```

Pie chart

to represent data in percentages

```python
# pie chart
data=['loan','education','entertainment','rental']
values=[10,20,30,40]
plt.pie(values,labels=data)
```

```python
e=[0.3,0,0,0]
plt.pie(values,labels=data,explode=e)
```
